## **Spam Classifier**

### **`Data Collection`**

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/SpamClassifier/master/smsspamcollection/SMSSpamCollection",sep='\t',names=['label','message'])
df
```

### **`Preprocessing`**

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()

lemma
```

```python
Pure_sentence = []
```

```python
for i in df['message']:

  sentence = re.sub('[^a-zA-Z]'," ",i)
  sentence = sentence.lower()
  sentence = sentence.split()
  sentence = [lemma.lemmatize(i) for i  in sentence if i not in stopwords.words('english')]
  sentence = ' '.join(sentence)
  Pure_sentence.append(sentence)
```

```python
len(Pure_sentence)
```

**`Text Into Numerical values`**

```python
# X Numericals
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer()

x = vector.fit_transform(Pure_sentence).toarray()

X = pd.DataFrame(x)
```

```python
X
```

```python
df['label'].unique()
```

```python
df['label'] = df['label'].map({'ham':0, 'spam':1})
```

```python
df['label']
```

**`Model Process`**

```python
y = df['label']
```

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test,y_pred)

acc
```
